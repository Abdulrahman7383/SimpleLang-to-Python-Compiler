```python
"""
مشروع المترجم البرمجي - Compiler Project
يتضمن جميع مراحل الترجمة: Lexer, Parser, Semantic Analyzer, Code Generator
"""

import re
import keyword
from dataclasses import dataclass, asdict
from typing import List, Tuple, Optional, Dict, Any


# ==================== 1. المحلل المعجمي (Lexer) ====================

@dataclass
class Token:
    type: str
    value: str
    line: int
    column: int
    file: Optional[str] = None

    def to_dict(self):
        return asdict(self)


class LexicalAnalyzer:
    """المحلل المعجمي - يحول الكود إلى رموز (Tokens)"""
    
    def __init__(self, code: str, filename: str = "<input>"):
        self.code = code
        self.filename = filename
        self.tokens: List[Token] = []
        self.errors: List[str] = []
        self.line = 1
        self.column = 1
        self.keywords = set(keyword.kwlist) | {'int', 'float', 'if', 'else', 'while', 'print'}
        
        # قواعس التحليل المعجمي
        self.rules = [
            ('COMMENT_SINGLE', r'//[^\n]*'),
            ('COMMENT_MULTI', r'/\*.*?\*/'),
            ('STRING', r'"[^"\\]*(\\.[^"\\]*)*"|\'[^\'\\]*(\\.[^\'\\]*)*\''),
            ('NUMBER_FLOAT', r'\d+\.\d+'),
            ('NUMBER', r'\d+'),
            ('ASSIGN', r'='), ('EQ', r'=='), ('NE', r'!='), 
            ('LT', r'<'), ('GT', r'>'),
            ('LE', r'<='), ('GE', r'>='), 
            ('PLUS', r'\+'), ('MINUS', r'-'), ('MUL', r'\*'),
            ('DIV', r'/'), ('LPAREN', r'\('), ('RPAREN', r'\)'),
            ('LBRACE', r'\{'), ('RBRACE', r'\}'),
            ('SEMICOLON', r';'),
            ('ID', r'[a-zA-Z_]\w*'), 
            ('NEWLINE', r'\n'), 
            ('SKIP', r'[ \t\r]+'), 
            ('MISMATCH', r'.'),
        ]
        
        pattern = '|'.join(f'(?P<{name}>{regex})' for name, regex in self.rules)
        self.regex = re.compile(pattern, re.DOTALL)

    def _update_position(self, text: str, start_pos: int):
        """تحديث موضع السطر والعمود"""
        lines_before = text[:start_pos].splitlines(keepends=True)
        self.line = len(lines_before)
        if lines_before:
            last_line_start = text.rfind('\n', 0, start_pos) + 1
            self.column = start_pos - last_line_start + 1
        else:
            self.column = start_pos + 1

    def analyze(self) -> Tuple[List[Token], List[str]]:
        """تحليل الكود وإرجاع الرموز والأخطاء"""
        pos = 0
        length = len(self.code)
        
        while pos < length:
            match = self.regex.match(self.code, pos)
            if not match:
                self.errors.append(f"خطأ معجمي: خطأ غير متوقع عند الموضع {pos}")
                break
                
            kind = match.lastgroup
            value = match.group()
            start_pos = pos
            pos = match.end()
            self._update_position(self.code, start_pos)

            # تخطي المسافات والتعليقات والأسطر الجديدة
            if kind in ('SKIP', 'NEWLINE', 'COMMENT_SINGLE', 'COMMENT_MULTI'):
                continue
                
            # التحقق من الكلمات المفتاحية
            if kind == 'ID' and value in self.keywords:
                kind = 'KEYWORD'
                
            # معالجة الرموز غير المعروفة
            if kind == 'MISMATCH':
                self.errors.append(f"خطأ معجمي: رمز غير معروف '{value}' في سطر {self.line}, عمود {self.column}")
                continue

            self.tokens.append(Token(kind, value, self.line, self.column, self.filename))
            
        return self.tokens, self.errors


# ==================== 2. المحلل النحوي (Parser) ====================

class Parser:
    """المحلل النحوي - يحول الرموز إلى شجرة بناء الجملة (AST)"""
    
    def __init__(self, tokens):
        self.tokens = tokens
        self.pos = 0
        self.current_token = self.tokens[self.pos] if tokens else None
        self.errors: List[str] = []

    def advance(self):
        """الانتقال إلى الرمز التالي"""
        self.pos += 1
        self.current_token = self.tokens[self.pos] if self.pos < len(self.tokens) else None

    def match(self, expected_type):
        """التحقق من نوع الرمز الحالي"""
        if self.current_token and self.current_token.type == expected_type:
            self.advance()
        else:
            actual = self.current_token.type if self.current_token else "EOF"
            error_msg = f"خطأ نحوي: متوقع {expected_type} ولكن وجد {actual}"
            self.errors.append(error_msg)
            raise SyntaxError(error_msg)

    def parse(self):
        """تحليل الكود الكامل"""
        statements = []
        try:
            while self.current_token:
                statements.append(self.statement())
        except SyntaxError:
            pass
        return statements

    def statement(self):
        """معالجة جملة واحدة"""
        tok = self.current_token
        if tok.type == 'KEYWORD' and tok.value == 'int':
            return self.declaration()
        elif tok.type == 'ID':
            return self.assignment()
        elif tok.type == 'KEYWORD' and tok.value == 'print':
            return self.print_stmt()
        elif tok.type == 'KEYWORD' and tok.value == 'if':
            return self.if_stmt()
        elif tok.type == 'KEYWORD' and tok.value == 'while':
            return self.while_stmt()
        elif tok.type == 'LBRACE':
            return self.block()
        else:
            raise SyntaxError(f'خطأ نحوي: جملة غير معروفة: {tok.value}')

    def declaration(self):
        """معالجة تعريف متغير"""
        self.match('KEYWORD')  # int
        var_name = self.current_token.value
        self.match('ID')
        self.match('ASSIGN')
        expr = self.expression()
        self.match('SEMICOLON')
        return ('declaration', var_name, expr)

    def assignment(self):
        """معالجة إسناد قيمة لمتغير"""
        var_name = self.current_token.value
        self.match('ID')
        self.match('ASSIGN')
        expr = self.expression()
        self.match('SEMICOLON')
        return ('assignment', var_name, expr)

    def print_stmt(self):
        """معالجة جملة الطباعة"""
        self.match('KEYWORD')  # print
        self.match('LPAREN')
        expr = self.expression()
        self.match('RPAREN')
        self.match('SEMICOLON')
        return ('print', expr)

    def if_stmt(self):
        """معالجة جملة الشرط"""
        self.match('KEYWORD')
        self.match('LPAREN')
        cond = self.condition()
        self.match('RPAREN')
        if_block = self.block()
        else_block = None
        if self.current_token and self.current_token.type == 'KEYWORD' and self.current_token.value == 'else':
            self.match('KEYWORD')
            else_block = self.block()
        return ('if', cond, if_block, else_block)

    def while_stmt(self):
        """معالجة حلقة while"""
        self.match('KEYWORD')
        self.match('LPAREN')
        cond = self.condition()
        self.match('RPAREN')
        body = self.block()
        return ('while', cond, body)

    def condition(self):
        """معالجة الشرط"""
        left = self.expression()
        op = self.current_token.type if self.current_token else None
        
        if op not in ('EQ', 'NE', 'LT', 'GT', 'LE', 'GE'):
            raise SyntaxError(f"خطأ نحوي: عامل مقارنة متوقع، وجد {op}")
        
        self.match(op)
        right = self.expression()
        return ('condition', left, op, right)

    def expression(self):
        """معالجة التعبير"""
        node = self.term()
        while self.current_token and self.current_token.type in ('PLUS', 'MINUS'):
            op = self.current_token.type
            self.advance()
            node = ('binop', op, node, self.term())
        return node

    def term(self):
        """معالجة الحد"""
        node = self.factor()
        while self.current_token and self.current_token.type in ('MUL', 'DIV'):
            op = self.current_token.type
            self.advance()
            node = ('binop', op, node, self.factor())
        return node

    def factor(self):
        """معالجة العامل"""
        tok = self.current_token
        if tok.type == 'NUMBER':
            self.advance()
            return ('number', float(tok.value) if '.' in tok.value else int(tok.value))
        elif tok.type == 'STRING':
            self.advance()
            return ('string', tok.value.strip('"\''))
        elif tok.type == 'ID':
            self.advance()
            return ('id', tok.value)
        elif tok.type == 'LPAREN':
            self.advance()
            node = self.expression()
            self.match('RPAREN')
            return node
        else:
            raise SyntaxError(f'خطأ نحوي: عامل غير متوقع: {tok.value}')

    def block(self):
        """معالجة كتلة من الجمل"""
        self.match('LBRACE')
        stmts = []
        while self.current_token and self.current_token.type != 'RBRACE':
            stmts.append(self.statement())
        self.match('RBRACE')
        return ('block', stmts)


# ==================== 3. المحلل الدلالي (Semantic Analyzer) ====================

class SemanticAnalyzer:
    """المحلل الدلالي - يتحقق من صحة المتغيرات والأنواع"""
    
    def __init__(self):
        self.symbols = set()
        self.errors: List[str] = []

    def analyze(self, ast):
        """تحليل شجرة البناء"""
        for node in ast:
            self.visit(node)
        return self.errors

    def visit(self, node):
        """زيارة عقدة في الشجرة"""
        if not isinstance(node, tuple) or len(node) == 0:
            return
        
        method_name = f'visit_{node[0]}'
        visitor = getattr(self, method_name, self.generic_visit)
        return visitor(node)

    def generic_visit(self, node):
        """معالجة عامة للعقد"""
        for item in node:
            if isinstance(item, tuple):
                self.visit(item)

    def visit_declaration(self, node):
        """معالجة تعريف متغير"""
        var_name = node[1]
        if var_name in self.symbols:
            self.errors.append(f"خطأ دلالي: إعادة تعريف المتغير '{var_name}'")
        self.symbols.add(var_name)
        self.visit(node[2])

    def visit_assignment(self, node):
        """معالجة إسناد قيمة"""
        var_name = node[1]
        if var_name not in self.symbols:
            self.errors.append(f"خطأ دلالي: المتغير '{var_name}' مستخدم قبل تعريفه!")
        self.visit(node[2])

    def visit_id(self, node):
        """معالجة معرّف"""
        if node[1] not in self.symbols:
            self.errors.append(f"خطأ دلالي: المتغير '{node[1]}' غير معرّف!")

    def visit_block(self, node):
        """معالجة كتلة"""
        for stmt in node[1]:
            self.visit(stmt)

    def visit_print(self, node):
        """معالجة جملة الطباعة"""
        self.visit(node[1])

    def visit_if(self, node):
        """معالجة جملة الشرط"""
        self.visit(node[1])
        self.visit(node[2])
        if node[3]:
            self.visit(node[3])

    def visit_while(self, node):
        """معالجة حلقة while"""
        self.visit(node[1])
        self.visit(node[2])

    def visit_condition(self, node):
        """معالجة الشرط"""
        self.visit(node[1])
        self.visit(node[3])

    def visit_binop(self, node):
        """معالجة العملية الثنائية"""
        self.visit(node[2])
        self.visit(node[3])


# ==================== 4. مولد الكود (Code Generator) ====================

class CodeGenerator:
    """مولد الكود - يحول AST إلى كود Python"""
    
    def __init__(self):
        self.code_lines: List[str] = []
        self.indent_level = 0

    def generate(self, ast):
        """توليد الكود"""
        self.code_lines = []
        self.indent_level = 0
        for node in ast:
            self.visit(node)
        return '\n'.join(self.code_lines)

    def add_line(self, line: str):
        """إضافة سطر كود"""
        indent = '    ' * self.indent_level
        self.code_lines.append(f"{indent}{line}")

    def visit(self, node):
        """زيارة عقدة"""
        if not isinstance(node, tuple) or len(node) == 0:
            return
        
        method_name = f'gen_{node[0]}'
        generator = getattr(self, method_name, self.generic_gen)
        return generator(node)

    def generic_gen(self, node):
        """معالجة عامة"""
        pass

    def gen_declaration(self, node):
        """توليد تعريف متغير"""
        self.add_line(f"{node[1]} = {self.visit_expr(node[2])}")

    def gen_assignment(self, node):
        """توليد إسناد قيمة"""
        self.add_line(f"{node[1]} = {self.visit_expr(node[2])}")

    def gen_print(self, node):
        """توليد جملة الطباعة"""
        self.add_line(f"print({self.visit_expr(node[1])})")

    def gen_if(self, node):
        """توليد جملة الشرط"""
        self.add_line(f"if {self.visit_expr(node[1])}:")
        self.indent_level += 1
        for stmt in node[2][1]:
            self.visit(stmt)
        self.indent_level -= 1
        if node[3]:
            self.add_line("else:")
            self.indent_level += 1
            for stmt in node[3][1]:
                self.visit(stmt)
            self.indent_level -= 1

    def gen_while(self, node):
        """توليد حلقة while"""
        self.add_line(f"while {self.visit_expr(node[1])}:")
        self.indent_level += 1
        for stmt in node[2][1]:
            self.visit(stmt)
        self.indent_level -= 1

    def gen_condition(self, node):
        """توليد نص الشرط"""
        left = self.visit_expr(node[1])
        op = node[2]
        right = self.visit_expr(node[3])
        
        op_map = {'EQ': '==', 'NE': '!=', 'LT': '<', 'GT': '>', 'LE': '<=', 'GE': '>='}
        return f"{left} {op_map.get(op, op)} {right}"

    def gen_block(self, node):
        """توليد كتلة"""
        for stmt in node[1]:
            self.visit(stmt)

    def visit_expr(self, node):
        """زيارة تعبير"""
        if not isinstance(node, tuple):
            return str(node)
        
        if node[0] == 'number':
            return str(node[1])
        elif node[0] == 'string':
            return f'"{node[1]}"'
        elif node[0] == 'id':
            return node[1]
        elif node[0] == 'binop':
            left = self.visit_expr(node[2])
            right = self.visit_expr(node[3])
            op_map = {'PLUS': '+', 'MINUS': '-', 'MUL': '*', 'DIV': '/'}
            op = op_map.get(node[1], node[1])
            return f"({left} {op} {right})"
        elif node[0] == 'condition':
            return self.gen_condition(node)
        else:
            return str(node)


# ==================== 5. فئة المترجم الرئيسية ====================

class Compiler:
    """فئة المترجم الرئيسية - تدير جميع مراحل الترجمة"""
    
    def __init__(self):
        self.lexer_errors = []
        self.parser_errors = []
        self.semantic_errors = []
        self.tokens = []
        self.ast = []
        self.generated_code = ""

    def compile(self, code: str) -> Dict[str, Any]:
        """تجميع الكود عبر جميع المراحل"""
        result = {
            "success": False,
            "error": None,
            "stage": None,
            "stageLabel": None,
            "stages": {
                "lexer": {"tokens": [], "errors": []},
                "parser": {"ast": [], "errors": []},
                "semantic": {"errors": []},
                "codegen": {"code": ""}
            }
        }

        try:
            # المرحلة 1: التحليل المعجمي
            lexer = LexicalAnalyzer(code)
            self.tokens, self.lexer_errors = lexer.analyze()
            result["stages"]["lexer"]["tokens"] = [t.to_dict() for t in self.tokens]
            result["stages"]["lexer"]["errors"] = self.lexer_errors

            if self.lexer_errors:
                result["success"] = False
                result["error"] = self.lexer_errors[0] if self.lexer_errors else "خطأ معجمي"
                result["stage"] = "lexer"
                result["stageLabel"] = "خطأ معجمي"
                return result

            # المرحلة 2: التحليل النحوي
            parser = Parser(self.tokens)
            self.ast = parser.parse()
            self.parser_errors = parser.errors
            result["stages"]["parser"]["ast"] = self._ast_to_dict(self.ast)
            result["stages"]["parser"]["errors"] = self.parser_errors

            if self.parser_errors:
                result["success"] = False
                result["error"] = self.parser_errors[0] if self.parser_errors else "خطأ نحوي"
                result["stage"] = "parser"
                result["stageLabel"] = "خطأ نحوي"
                return result

            # المرحلة 3: التحليل الدلالي
            semantic = SemanticAnalyzer()
            self.semantic_errors = semantic.analyze(self.ast)
            result["stages"]["semantic"]["errors"] = self.semantic_errors

            if self.semantic_errors:
                result["success"] = False
                result["error"] = self.semantic_errors[0] if self.semantic_errors else "خطأ دلالي"
                result["stage"] = "semantic"
                result["stageLabel"] = "خطأ دلالي"
                return result

            # المرحلة 4: توليد الكود
            codegen = CodeGenerator()
            self.generated_code = codegen.generate(self.ast)
            result["stages"]["codegen"]["code"] = self.generated_code

            result["success"] = True
            return result

        except Exception as e:
            result["success"] = False
            result["error"] = str(e)
            return result

    def _ast_to_dict(self, node):
        """تحويل AST إلى قاموس قابل للتسلسل"""
        if isinstance(node, list):
            return [self._ast_to_dict(item) for item in node]
        elif isinstance(node, tuple):
            return [self._ast_to_dict(item) for item in node]
        else:
            return node
            # ==================== 6. التشغيل الرئيسي ====================
if __name__ == "__main__":
    # اقرأ الكود من ملف code.txt
    try:
        with open("code.txt", "r", encoding="utf-8") as f:
            code = f.read()
    except FileNotFoundError:
        print("الملف code.txt غير موجود. سيتم استخدام مثال افتراضي.")
        code = """
int x = 10;
int y = 20;
print(x + y);
"""
    
    print("=" * 50)
    print("الكود المدخل:")
    print(code)
    print("=" * 50)
    
    # إنشاء مترجم
    compiler = Compiler()
    result = compiler.compile(code)
    
    # عرض النتيجة
    if result["success"]:
        print("\n✅ الترجمة ناجحة!")
        print("\nالكود المولد (Python):")
        print(result["stages"]["codegen"]["code"])
        
        # حفظ الكود المولد
        with open("output.py", "w", encoding="utf-8") as f:
            f.write(result["stages"]["codegen"]["code"])
        print("\n🎉 تم إنشاء output.py. شغله باستخدام: python output.py")
    else:
        print(f"\n❌ فشلت الترجمة في مرحلة {result['stageLabel']}")
        print(f"الخطأ: {result['error']}")
```
